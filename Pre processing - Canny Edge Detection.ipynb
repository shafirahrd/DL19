```python
from matplotlib import pyplot as plt
from scipy import misc
from scipy import ndimage
import numpy as np
import cv2

img = cv2.imread("datasetIris/001/1/001_1_2.bmp")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
img_blurred = ndimage.gaussian_filter(img, sigma=1.4)

plt.imshow(img_blurred, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
edges = cv2.Canny(img_blurred, 100, 200)

plt.imshow(edges, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, 1, img.shape[0]/64, param1=50, param2=30, minRadius=0, maxRadius=0)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        # Draw outer circle
        cv2.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 2)
        # Draw inner circle
        cv2.circle(img, (i[0], i[1]), 2, (0, 0, 255), 3)
        
plt.imshow(img)
plt.show()
```
